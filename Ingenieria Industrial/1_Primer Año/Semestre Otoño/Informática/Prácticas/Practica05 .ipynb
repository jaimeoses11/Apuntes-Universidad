```python
# Este programa llama a varias funciones que introduciendo 2 listas de números el primer elemento coincidente,
# una suma según ciertas condiciones y la mayor diferencia entre elementos de las 2 listas.
N = int (input ('Introduce el número de elementos de cada lista:'))
def lee_lista (y):
    x = input ('Introduce la {} lista de {} números separados por espacios:'.format(y,N))
    x = x.split ()
    for i in range (N):
        x[i]=int(x[i])
    return x

def posicion_coincidentes (x, y):
    posicion = -1
    for i in range (N):
        if x[i]==y[i]:
            posicion = i
            break
    print ('Los primeros elementos coincidentes están en la posición:', posicion)

def suma (x, y):
    suma = 0
    for i in range (N):
        if x[i]>y[i]:
            suma += 1
        elif x[i]<y[i]:
            suma -= 1
        elif x[i]==y[i]:
            suma+=0
    print ('La suma solicitada es:', suma) 

def maxima_diferencia (x, y):
    mayor = 0
    for i in range (N):
        for j in range (N):
            dif = x[i]-y[j]
            if dif<0:
                dif*=(-1)
            if mayor<dif:
                mayor = dif
    print ('La mayor diferencia entre elementos de a y b es:', mayor)

a = lee_lista ('primera')
b = lee_lista ('segunda')
posicion_coincidentes (a, b)
suma (a, b)
maxima_diferencia (a, b)  
```

```python
# Este programa llama a varias funciones que, a apartir de una matriz y una lista, dan una nueva lista en la que el elemento
# 'i' será el resultado de sumar los elementos de m correspondientes a la columna v[i] (de la lista introducida) y restar
# los de la fila v[i] (de la lista introducida) y por ultimo da la suma de todos los elementos de la matriz.
N = int (input ('Introduce el número de filas y columnas de la matriz:'))
def lee_matriz ():
    x = []
    for i in range (N):
        y = input ('Escribe la fila {} de la matriz {}x{} separada por espacios:'.format(i, N, N))
        y = y.split()
        x.append (y)
    for i in range (N):
        for j in range (N):
            x[i][j] = int(x[i][j])
    return x

def lee_lista ():
    v = (input ('Introduce una lista separada por espacios:'))
    v = v.split ()
    for i in range (len(v)):
        v[i]=int(v[i])
    return v

def cruz (x, y):
    suma = 0
    cruz = 0
    t = []
    for a in range (N):
        for i in range (N):
            for j in range (N):
                if j==y[a]:
                    cruz+=x[i][j]
                if i==y[a]:
                    cruz-=x[i][j]
        t.append (cruz)
        cruz = 0
    for i in range (N):
        for j in range (N):
            suma += x[i][j]
    print ('La nueva lista es {} y la suma total es {}'.format(t, suma))

m = lee_matriz ()
v = lee_lista ()
cruz (m, v)
```
