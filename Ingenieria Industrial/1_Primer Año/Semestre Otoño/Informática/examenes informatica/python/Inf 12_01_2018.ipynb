```python
### Ejercicio 2

def LeeMatriz():
    print('Vamos a introducir una matriz',N,'x',M)
    m = []
    for i in range(N):
        fila = input('Mete la fila {}: '.format(i)).split()
        for j in range(M):
            fila[j] = int(fila[j])
        m.append(fila)
    return m

def SubMatriz(m, f, c):
    mSub = []
    for i in range(N):
        mSub.append([0] * M)
    for i in range(f + 1):
        for j in range(c + 1):
            mSub[i][j] = m[i][j]
    return mSub

def SumaMatriz(m):
    suma = 0
    for fila in m:
        suma += sum(fila)
    return suma

def Sumas(m):
    mAc = []
    for i in range(N):
        mAc.append([0] * M)
    for i in range(N):
        for j in range(M):
            mAc[i][j] = SumaMatriz(SubMatriz(m, i, j))
    return mAc

# PROGRAMA PRINCIPAL ##################

N = 4
M = 5
mOr = LeeMatriz()
mRes = Sumas(mOr)
print('La suma de los valores de la matriz acumulada es:',SumaMatriz(mRes))
```
